```python
# 关键词列表
keywords = [""]
```

```python
import datetime
import sys
import os
import codecs
```

```python
keywords = []
stopwords = []

with codecs.open("../data/dict/keywords.dict", "r", "utf-8") as f:
    for word in f:
        if len(word) > 1:
            keywords.append(word[:-1])
with codecs.open("../data/dict/stopwords.dict", "r", "utf-8") as f:
    for word in f:
        if len(word) > 1:
            stopwords.append(word[:-1])
            
print(keywords)
print(stopwords)
```

```python
texts = codecs.open("../data/announcement/tingpai-test.txt", "r", "utf-8")

for line in texts:
    items = line.split("_")
    code = items[0]
    title = items[1]
    datestr = items[2]
    timestr = ":".join(items[3:5]) + ":" + items[5][:2]
    content = items[5][2:]
    
    print(code, title, datestr, timestr)
    
    
    
    print("[摘要]", extract_summary(content, '。')) 
    print("[原文]", content)

texts.close()
```

```python
# 抽取摘要
def extract_summary(content, split_flag):
    # 先进行分句，使用 split_flag（一般是逗号或者句号）
    sentences = content.split(split_flag)
    print("原文共", len(sentences), "个分句")
    result = []
    
    for s in sentences:
        skip = False
        for word in stopwords:
            if word in s:
                skip = True
                break
                
        # 判断句子中是否包含关键词
        if skip:
            continue
        for word in keywords:
            if word in s:
                result.append(s)
                break
    print("摘要共", len(result), "个分句")
    return split_flag.join(result)
```
